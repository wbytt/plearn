### Data Strcuture
we have 4 kinds of data : `set`,`tuple`,`list`,`dictionary`
here are some basic knowledge:
1. list: the same as arrays, but we can access reversely, `list[-1]` can only restore the same kind of data.  ` ['a','b','c'] `
2. tuple: similar to list ,but the data is unchangable, but can store different kinds of data(restore pointer). ` ('a',1,[1,2]) `
3. dictionary: a list plus,can restore any kinds of data,and append freely, use key to access data(dont have order),just a pointer to various data. `{'a':1,'b':'c'}`
4. set:no order,cant access, can only restore,compare,operate specified data. `{1,2,3} or set([1,4])`

### Compression
a kind of method to construct data structure, it just use a data structure to build another   

````python
# list,tuple,set
print([i for i in range(5) if i % 3 ==0])
print(tuple(i for i in range(5) if i % 3 ==0))
print({i for i in range(5) if i % 3 ==0})
# dictionary
key=['a','b','c']
print({c:i for c,i in zip(key,range(5))})# only get the aligned 3 data
````

### interator and generator
a tool used to traverse the dataset one by one,and cant reverse     
and a similar one that reture a interator   
usually we use generator to generate a interator, and use it to access the required data.

````python
# interator: perform well when diy the rules of interate
list=[1,2,3,4]
it=iter(list)
for i in it:
    print(i,end=" ")
# generator: usually use it in functions,return a object of interator(iter)(set of data)
# it generate numbers after used(take less memory)
def iter_1(n):
    for i in range(n):
        yield i
gen=iter_1(10000) #return a iter, save much memory
print(next(gen))
````

### errors and exceptions
usually we use try/except to judge if the code is right

````python
# judge if the operation is legal
try:
    x=int(input())
except ValueError:
    print("input error") #if it is do
else:
    print("input accepted") # if not do
finally:
    print(x) #no matter what do
# in some case intrige error
if x>10:
    raise Exception('Error') # similar to print
````
